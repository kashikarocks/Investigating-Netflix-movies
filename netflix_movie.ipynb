```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
nineties_movie= netflix_df[(netflix_df['release_year']<2000) & (netflix_df['release_year']>=1990)]
```

```python
print(nineties_movie)
```

# The most frequent movie duration in the 1990's(use 1990 as the decade's start year).

```python
duration = nineties_movie['duration'].mode()[0]
```

```python
print(duration)
```

# Visualization of most frequent movie duration in the 1990's

```python
plt.hist(nineties_movie['duration'],edgecolor='black', alpha=0.7)
plt.xlabel('Duration')
plt.ylabel('freq')
plt.title('Distribution of movie durations in 1990s')
plt.show()
```

A movie is considered short if it is less than 90 minutes.

```python
shrt_mv = nineties_movie[nineties_movie['duration']<90]
```

```python
print(shrt_mv)
```

#  Count the number of short action movies released in the 1990s 

```python
sm = shrt_mv[shrt_mv['genre']=='Action']
```

```python
short_movie_count = int(sm.shape[0])
```
